# Insurance Policy Distribution Analysis by Area

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

## Number of Policies per Area

```python
area_counts = df['Area'].value_counts()
area_counts.plot(kind='bar')
plt.ylabel('Number of Policies')
plt.show()
```

## Policy Type Distribution

```python
pd.crosstab(df['Area'], df['Policy_Type']).plot(kind='bar', stacked=True)
plt.show()
```

## Premium Revenue by Area

```python
premium_area = df.groupby('Area')['Premium_Amount'].sum()
premium_area.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```
